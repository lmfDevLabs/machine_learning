```python
# predict.py
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model

def load_and_predict(input_csv, model_path, output_txt):
    # Carga el modelo
    model = load_model(model_path)

    # Carga los datos para predicci칩n
    data = pd.read_csv(input_csv)
    X = data.values  # Asumiendo que 'data' ya est치 preprocesado adecuadamente

    # Realiza las predicciones
    predictions = model.predict(X)
    predicted_labels = (predictions > 0.5).astype(int)  # Asumiendo clasificaci칩n binaria

    # Guarda las predicciones en un archivo de texto
    np.savetxt(output_txt, predicted_labels, fmt='%d')

if __name__ == "__main__":
    input_csv = 'quiz.csv'  # El path al CSV de entrada
    model_path = 'best_model.h5'  # El path al modelo H5 guardado
    output_txt = 'answers.txt'  # El path al archivo de texto de salida
    load_and_predict(input_csv, model_path, output_txt)
```
